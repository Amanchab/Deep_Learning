We will be using Haar wavelet transform

```python
import cv2
```

```python
alg = 'data/haarcascade_frontalface_default.xml'
haar_cascade = cv2.CascadeClassifier(alg)
cam = cv2.VideoCapture(0)
```

```python
while True:
    _,img = cam.read()
    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # source, scale factor (it is a trained model so we shouldn't give the size more than what is in the trained model), minimum neighbours ()
    face = haar_cascade.detectMultiScale(grayImg, 1.3, 4)
    for (x, y, w, h) in face:
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)

        cv2.imshow("FaceDetection", img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break
    print('Image Captured successfully')
```

```python
cam.release()
cv2.destroyAllWindows()
```
